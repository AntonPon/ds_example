```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_path = '../data/train.csv'
```

```python
train_df = pd.read_csv(train_path)
train_df.shape
```

There are in total 53 data variables and one target variable. 42 variables of type float and 11 of type integer.  There are 90000 examples. Thre is no missing data. Variable 8 looks like categorical, hast to be investigated further.

```python
train_df.describe()
```

```python
numeric = list(train_df.select_dtypes("float64").columns)
numeric.remove('target')
print('total numeric features:', len(numeric))
cat = list(train_df.select_dtypes("int64").columns)
print('numeric_features:\n', numeric)
print('total integer features:', len(cat))
print('integer_features:\n', cat)
for i in cat:
    print(f'{np.round((train_df[i].value_counts() / train_df[i].value_counts().sum()) * 100,2)}')
```

In total, there are 53 variables, 52 of which are continuous, and 1 is categorical. The categorical one is varible 8. Also, the variable 6 is uniformly distributed. To use this variable for training it has to be splitted into two variables. 

```python
fig, ax = plt.subplots(4,3,figsize=(35,35))
row = col = 0
for n, i in enumerate(cat):
    if (n%3 == 0) & (n > 0):
        row += 1
        col = 0
    train_df[i].plot(kind="kde",ax=ax[row,col])
    ax[row,col].set_title(i)
    col += 1
```

By inspecting Kernel Density Estimation plots, integer type variables except number 8 are uniformly distributed. This will influence the normalization and model training procedure.

```python
fig, ax = plt.subplots(6,7,figsize=(35,35))
row = col = 0
for n, i in enumerate(numeric):
    if (n%7 == 0) & (n > 0):
        row += 1
        col = 0
    train_df[i].plot(kind="kde",ax=ax[row,col])
    ax[row,col].set_title(i)
    col += 1
```

By inspecting Kernel Density Estimation plots for numeric features, we observe that variable 6 is bimodally distributed, and all other numeric features are uniformly distributed.
